```python
import pandas as pd 
df = pd.read_csv("HousingData.csv")
df
```

```python
# We need to check the null values
df = df.dropna()
df.isnull().sum()
# as there are no null values in the dataset we can move ahead
```

```python
df
```

```python
X = df.drop(columns = ["MEDV"])
X
```

```python
y = df["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
y_pred_train = lr.predict(X_train)
y_pred_train
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train, y_pred_train)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_train, y_pred_train) # (0 - 1)
# higher the r2 score, better the model is
```

```python
# now we will test the data it has never seen the test data
y_pred_test = lr.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred_test)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.show()
```

```python
r2 = r2_score(y_test, y_pred_test)
print("r2 Score : ", r2)
```

```python
# The error should be as low as possible
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred_test)
mae = mean_absolute_error(y_test,y_pred_test)
print("Mean Square Error : ", mse)
print("Mean Absolute Error : ", mae)
```
